```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
df= pd.read_csv('TravelMode.csv')
```

```python
choice_air = len(df.loc[(df['mode'].str.contains('air')) & (df['choice'] == 'yes')])
choice_train = len(df.loc[(df['mode'].str.contains('train')) & (df['choice'] == 'yes')])
choice_bus = len(df.loc[(df['mode'].str.contains('bus')) & (df['choice'] == 'yes')])
choice_car = len(df.loc[(df['mode'].str.contains('car')) & (df['choice'] == 'yes')])
```

```python
values=[choice_air, choice_train, choice_bus, choice_car]
labels= ['air', 'train', 'bus', 'car']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1.2,0.3))
plt.show
```

```python
pd.options.display.float_format = '{:.2f}'.format
df= pd.read_csv('TravelMode.csv')
df['time loss']=np.log(df['wait']+df['travel'])
df['e_cost']=(df['vcost']+df['gcost']) / df['income']
df
```

```python
data_air_no = df.loc[df['mode'].str.contains('air') & (df['choice'] == 'no')]
data_train_no = df.loc[df['mode'].str.contains('train') & (df['choice'] == 'no')]
data_bus_no = df.loc[df['mode'].str.contains('bus') & (df['choice'] == 'no')]
data_car_no = df.loc[df['mode'].str.contains('car') & (df['choice'] == 'no')]
data_air_yes = df.loc[df['mode'].str.contains('air') & (df['choice'] == 'yes')]
data_train_yes = df.loc[df['mode'].str.contains('train') & (df['choice'] == 'yes')]
data_bus_yes = df.loc[df['mode'].str.contains('bus') & (df['choice'] == 'yes')]
data_car_yes = df.loc[df['mode'].str.contains('car') & (df['choice'] == 'yes')]
```

```python
fig, axs= plt.subplots(2,2, figsize=(15,10))
fig.suptitle('Time Loss - Cost for each mode')

#1
axs[0,0].scatter(data_air_no['time loss'], data_air_no['e_cost'], c='blue', label='no')
axs[0,0].scatter(data_air_yes['time loss'], data_air_yes['e_cost'], c='skyblue', label='yes')
axs[0,0].set_title('Air')
axs[0,0].set(xlabel='ln(Time Loss)', ylabel='Cost / Income')
axs[0,0].legend()
axs[0,0].set_xlim(4, 8)
axs[0,0].set_ylim(0, 75)


#2
axs[0,1].scatter(data_train_no['time loss'], data_train_no['e_cost'], c='orange', label='no')
axs[0,1].scatter(data_train_yes['time loss'], data_train_yes['e_cost'], c='navajowhite', label='yes')
axs[0,1].set_title('Train')
axs[0,1].set(xlabel='ln(Time Loss)', ylabel='Cost / Income')
axs[0,1].legend()
axs[0,1].set_xlim(4, 8)
axs[0,1].set_ylim(0, 75)

#3
axs[1,0].scatter(data_bus_no['time loss'], data_bus_no['e_cost'], c='green', label='no')
axs[1,0].scatter(data_bus_yes['time loss'], data_bus_yes['e_cost'], c='springgreen', label='yes')
axs[1,0].set_title('Bus')
axs[1,0].set(xlabel='ln(Time Loss)', ylabel='Cost / Income')
axs[1,0].legend()
axs[1,0].set_xlim(4, 8)
axs[1,0].set_ylim(0, 75)

#4
axs[1,1].scatter(data_car_no['time loss'], data_car_no['e_cost'], c='red', label='no')
axs[1,1].scatter(data_car_yes['time loss'], data_car_yes['e_cost'], c='darksalmon', label='yes')
axs[1,1].set_title('Car')
axs[1,1].set(xlabel='ln(Time Loss)', ylabel='Cost / Income')
axs[1,1].legend()
axs[1,1].set_xlim(4, 8)
axs[1,1].set_ylim(0, 75)
```
